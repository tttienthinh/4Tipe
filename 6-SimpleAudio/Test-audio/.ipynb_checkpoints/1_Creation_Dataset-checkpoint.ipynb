## Création d'audio

```python
def AUDIO_HTML(message="Tu as 1,5 secondes pour parler !!!"):
    return """
<html>
<div style="text-align: center;">
    <h1>
        Agathe & 4T
    </h1>
    <p>"""+message+"""</p>
    <button id="button">Click to record</button>
</div>
<script>
    var recordButton = document.getElementById("button")

    var base64data = 0;
    var reader;
    var recorder, gumStream;

    var handleSuccess = function (stream) {
        gumStream = stream;
        var options = {
            //bitsPerSecond: 8000, //chrome seems to ignore, always 48k
            mimeType: 'audio/webm;codecs=opus'
        };
        recorder = new MediaRecorder(stream);
        recorder.ondataavailable = function (e) {
            

            reader = new FileReader();
            reader.readAsDataURL(e.data);
            reader.onloadend = function () {
                base64data = reader.result;
            }
        };
    };
    navigator.mediaDevices.getUserMedia({ audio: true }).then(handleSuccess);

    function startRecord() {
            recorder.start();
            sleep(1500).then(() => {
                toggleRecording()
            })
    }
    function toggleRecording() {
        if (recorder && recorder.state == "recording") {
            recorder.stop();
            gumStream.getAudioTracks()[0].stop();
            recordButton.innerText = "Saving the recording... pls wait!"
        } else {
            recordButton.innerText = "Recording...";
            sleep(150).then(() => {
                startRecord()
            })
        }
    }
    function sleep(ms) {
        return new Promise(resolve => setTimeout(resolve, ms));
    }

    var data = new Promise(resolve => {;
        recordButton.onclick = () => {
            toggleRecording()
            sleep(2000).then(() => {
                resolve(base64data.toString())

            });
        }
    });
</script>
</html>
"""
```

```python
def get_audio(message="Tu as 1,5 secondes pour parler !!!"):
  display(HTML(AUDIO_HTML(message=message)))
  data = eval_js("data")
  binary = b64decode(data.split(',')[1])
  
  process = (ffmpeg
    .input('pipe:0')
    .output('pipe:1', format='wav')
    .run_async(pipe_stdin=True, pipe_stdout=True, pipe_stderr=True, quiet=True, overwrite_output=True)
  )
  output, err = process.communicate(input=binary)
  
  riff_chunk_size = len(output) - 8
  # Break up the chunk size into four bytes, held in b.
  q = riff_chunk_size
  b = []
  for i in range(4):
      q, r = divmod(q, 256)
      b.append(r)

  # Replace bytes 4:8 in proc.stdout with the actual size of the RIFF chunk.
  riff = output[:4] + bytes(b) + output[8:]

  sr, audio = wav_read(io.BytesIO(riff))

  return audio, sr
```

```python
audio, sr = get_audio()
plt.plot(audio)
plt.show()
Audio(data=audio, rate=48_000)
```

## Augmentation de données
https://medium.com/@makcedward/data-augmentation-for-audio-76912b01fdf6  


```python
import librosa
```

### Noise Injection

```python
def noise_manipulation(data, noise_factor=5_000):
    noise = np.random.randn(len(data))
    augmented_data = data + noise_factor * noise
    # Cast back to same data type
    augmented_data = augmented_data.astype(type(data[0]))
    return augmented_data
```

```python
result = noise_manipulation(audio, 100) # variant de 0 à 10_000
plt.plot(result, label="result")
plt.plot(audio, label="audio")
plt.legend()
Audio(data=result, rate=48_000)
```

### Shifting Time

```python
def shift_manipulation(data, sampling_rate, shift_max, shift_direction):
    shift = np.random.randint(max(sampling_rate * shift_max, 1))
    if shift_direction == 'right':
        shift = -shift
    elif shift_direction == 'both':
        direction = np.random.randint(0, 2)
        if direction == 1:
            shift = -shift
    augmented_data = np.roll(data, shift)
    # Set to silence for heading/ tailing
    n = len(data)
    cut = int((n - sampling_rate)/2)
    data_final = augmented_data[cut:-cut] # permet d'obtenir des son de 1 sec 48kbit
    return data_final
```

```python
result = shift_manipulation(audio, 48_000, 0.5, "both") # variant de 0 à 0.75
plt.plot(audio, label="audio")
plt.plot(result, label="result")
plt.legend()
Audio(data=result, rate=48_000)
```

### Changing Pitch

```python
def pitch_manipulate(data, sampling_rate, pitch_factor):
    return librosa.effects.pitch_shift(data.astype(np.float32), sampling_rate, pitch_factor)
```

```python
result = pitch_manipulate(audio, 48_000, -1.5) # variant de -2.5 à 2.5
plt.plot(audio, label="audio")
plt.plot(result, label="result")
plt.legend()
Audio(data=result, rate=48_000)
```

### Changing Speed

```python
def speed_manipulate(data, speed_factor):
    return librosa.effects.time_stretch(data.astype(np.float32), speed_factor)
```

```python
result = speed_manipulate(audio, .5) # variant de .5 à 1.5
plt.plot(audio, label="audio")
plt.plot(result, label="result")
plt.legend()
Audio(data=result, rate=48_000)
```

## Création de données

```python
import os

names = ["Bensal", "Chateau", "Gayout", "Le_Grandic", "Mistler", "Schuschu"]
```

```python
# Création des dossiers de sauvegarde RESULT
if not os.path.isdir("result"):
    os.mkdir("result")
for name in names:
    path = f"result/{name}"
    if not os.path.isdir(path):
        os.mkdir(path)
path = f"result/original"
if not os.path.isdir(path):
    os.mkdir(path)
```

```python
n = 1 # Nombre de données à enregistrer par classe
audios = {name:[] for name in names} # enregistrement des audios
for i in range(n):
    for name in names:
        audio, sr = get_audio(name)
        clear_output()
        Audio(data=audio, rate=48_000)
        audios[name].append(audio)
        save(audio, name=f"original/{name}{i}.wav")
```

```python
for name in names:
    new_audios = [] # PITCH
    for i in range(3): # 3+1 fois plus de données
        pitch = np.random.random()*5 - 2.5 # ptich varie entre [-2.5 et 2.5]
        for audio in audios[name]:
            new_audios.append(pitch_manipulate(audio, 48_000, pitch))
    audios[name] += new_audios.copy()

    new_audios = [] # SPEED
    for i in range(3): # 3+1 fois plus de données
        speed = np.random.random() + 0.5 # speed varie entre [0.5 et 1.5]
        for audio in audios[name]:
            new_audios.append(speed_manipulate(audio, speed))
    audios[name] += new_audios.copy()

    new_audios = [] # NOISE
    for noise_max in [100, 1_000, 10_000]: # 3+1 fois plus de données
        noise = np.random.randint(noise_max)
        for audio in audios[name]:
            new_audios.append(noise_manipulation(audio, noise_factor=noise))
    audios[name] += new_audios.copy()

    new_audios = [] # SHIFT
    for shift in range(4): # 4 fois plus de données
        for audio in audios[name]:
            new_audios.append(shift_manipulation(audio, 48_000, shift/4, "both")) # shift varie dans [0, 0.25, 0.5, 0.75]
    audios[name] = new_audios.copy()

    # 4 * 4 * 4 * 4 = 256
```

# Enregistrement des données

```python
# Enregistrement
for name in names:
    for i in range(len(audios[name])):
        audio = audios[name][i]
        save(audio, name=f"{name}/{i}.wav")
```
