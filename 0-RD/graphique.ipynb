```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x, derive=False):
    """
    Fonction Sigmoid
    """
    if derive:
        return np.exp(-x) / ((1+np.exp(-x)) ** 2)
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(12, 6))
plt.plot(x, sigmoid(x), label="Sigmoïde")
plt.grid()
plt.legend()
plt.savefig("1-trace sigmoid.jpg")
```
